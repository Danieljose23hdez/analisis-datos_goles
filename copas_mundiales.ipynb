```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
data = {
    "AÃ±o": [1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 
            1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018, 2022],
    "Goles_A": [4, 2, 4, 2, 3, 5, 3, 4, 4, 2, 3, 3, 3, 1, 0, 3, 2, 1, 1, 1, 4, 3],
    "Goles_B": [2, 1, 2, 1, 2, 2, 1, 2, 1, 1, 1, 1, 2, 0, 0, 0, 0, 1, 0, 0, 2, 3]
}

df = pd.DataFrame(data)
```

```python
df["Goles_A"] = pd.to_numeric(df["Goles_A"], errors="coerce")
df["Goles_B"] = pd.to_numeric(df["Goles_B"], errors="coerce")
df["Total_Goles"] = df["Goles_A"] + df["Goles_B"]
```

```python
promedio_goles = df["Total_Goles"].mean()
print(f"Promedio de goles por final: {promedio_goles:.2f}")
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["AÃ±o"], df["Total_Goles"], marker="o", linestyle="-", color="b", label="Goles por final")
plt.axhline(promedio_goles, color="r", linestyle="--", label=f"Promedio ({promedio_goles:.2f} goles)")
plt.xlabel("AÃ±o")
plt.ylabel("Total de Goles en la Final")
plt.title("Tendencia de Goles en Finales del Mundial")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df["AÃ±o"].values.reshape(-1, 1)
y = df["Total_Goles"].values

modelo = LinearRegression()
modelo.fit(X, y)

# Predecir para 2026
goles_2026 = modelo.predict([[2026]])[0]
print(f"PredicciÃ³n de goles para la final de 2026: {goles_2026:.2f}")
```
